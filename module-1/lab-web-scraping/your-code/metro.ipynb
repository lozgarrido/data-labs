```python
#scrapMetro

import requests
from pprint import pprint
from bs4 import BeautifulSoup

METRO_URL = "https://www.metromadrid.es/es/"

response = requests.get(METRO_URL)
soup = BeautifulSoup(response.text, "html.parser")

lines_container = soup.find("div", {"class": "small-12 medium-12 large-6 columns box__red-real"})
print(lines_container)
```

```python
zoomlines = lines_container.find_all("a", {"role":"link"})
print(zoomlines)
```

```python
lineslinks = [zoomline.get('href') for zoomline in zoomlines]
pprint(lineslinks)
```

```python
absolutUrls = [("https://www.metromadrid.es"+link) for link in lineslinks]
    
pprint(absolutUrls)
```
