# Challenge 1

#### Import all the libraries that are necessary.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
#1st way: different visualizations
plt.plot(x, y, color="b", ls=':')
plt.title("Double of numbers");
```

```python
plt.plot(x, z, color="k", ls='-')
plt.title("Square of numbers");
```

```python
#2nd way (subplots)

#Size
plt.subplots(figsize=(20,5))

#1. First subplot
plt.subplot(1,2,1)
#First serie
plt.plot(x, y)
#Enter title
plt.title("Double")

#2. Second subplot
plt.subplot(1,2,2)
#Second serie
plt.plot(x, z, ls=":")
#Enter title
plt.title("Square");
```

# 2nd way: join both visualizations

#### Augmenting your previous code, resize your previous plot.


*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
#Size
plt.figure(figsize=(20,5))

#Series
plt.plot(x, y)
plt.plot(x, z, ls=":")

#Title
plt.title("Square VS double");
```

#### Augmenting your previous code, label your axis.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
#Size
plt.figure(figsize=(20,5))

#Series
plt.plot(x, y)
plt.plot(x, z, ls=":")

#Axis labels
plt.xlabel("Variable")
plt.ylabel("Result")
#plt.xlim(a, b) and plt.ylim(a, b) to set limits

#Title
plt.title("Square VS double");
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
x = np.arange(0,10)
serie_1 = x**2
serie_2 = np.exp(x)

#Size
plt.figure(figsize=(20,10))

#Series
plt.plot(x, serie_1)
plt.plot(x, serie_2, ls=":")

#Axis and scales
plt.xlabel("Variable")
plt.xscale("linear")

plt.ylabel("Result")
plt.yscale("log")

#Title
plt.title("Square VS e^x");
```
